```python
def process_image(im, transforms = transformations):
    h, w, c = im.shape
    image = np.transpose(np.array(im), (2, 0, 1))
    image = np.expand_dims(image, axis=0)
    image = torch.tensor(image, dtype=torch.float32, device=DEVICE)
    if transforms is not None:
        image = transforms(image)
    image = image / 127.5 - 1
    return image
```

```python
def prepare_image(image):
    image = (image + 1) * 127.5
    return np.transpose(np.array(image.cpu(), dtype=np.int32).squeeze(0), (1, 2, 0))
```

## Creating the Generator model

```python
OUT_CHANNELS = 3
```

```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()

        # encoder (downsampling)
        self.enc_conv0 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=64, kernel_size=4, stride=2, padding=1, bias=False),
            nn.LeakyReLU()
        )
        self.enc_conv1 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU()
        )
        self.enc_conv2 = nn.Sequential(
            nn.Conv2d(in_channels=128, out_channels=256, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU()
        )
        self.enc_conv3 = nn.Sequential(
            nn.Conv2d(in_channels=256, out_channels=512, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU()
        )
        self.enc_conv4 = nn.Sequential(
            nn.Conv2d(in_channels=512, out_channels=512, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU()
        )
        self.enc_conv5 = nn.Sequential(
            nn.Conv2d(in_channels=512, out_channels=512, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU()
        )
        self.enc_conv6 = nn.Sequential(
            nn.Conv2d(in_channels=512, out_channels=512, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU()
        )
        self.enc_conv7 = nn.Sequential(
            nn.Conv2d(in_channels=512, out_channels=512, kernel_size=4, stride=2, padding=1, bias=False),
            nn.LeakyReLU()
        )        

        # decoder (upsampling)
        self.upsample0 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=512, out_channels=512, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.Dropout2d(0.5),
            nn.ReLU()
        )
        self.upsample1 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=1024, out_channels=512, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.Dropout2d(0.5),
            nn.ReLU()
        )
        self.upsample2 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=1024, out_channels=512, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.Dropout2d(0.5),
            nn.ReLU()
        )
        self.upsample3 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=1024, out_channels=512, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.ReLU()
        )
        self.upsample4 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=1024, out_channels=256, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU()
        )
        self.upsample5 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=512, out_channels=128, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )
        self.upsample6 = nn.Sequential(
            nn.ConvTranspose2d(in_channels=256, out_channels=64, kernel_size= 4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU()
        )
        
        self.out = nn.ConvTranspose2d(in_channels=128, out_channels=OUT_CHANNELS, kernel_size=4, stride=2, padding=1)
        self.act = nn.Tanh()

    def forward(self, x):
        # encoder
        e0 = self.enc_conv0(x)
        e1 = self.enc_conv1(e0)
        e2 = self.enc_conv2(e1)
        e3 = self.enc_conv3(e2)
        e4 = self.enc_conv4(e3)
        e5 = self.enc_conv5(e4)
        e6 = self.enc_conv6(e5)
        e7 = self.enc_conv7(e6)

        # decoder
        up0 = torch.cat((self.upsample0(e7), e6.clone()), dim=1)
        up1 = torch.cat((self.upsample1(up0), e5.clone()), dim=1)
        up2 = torch.cat((self.upsample2(up1), e4.clone()), dim=1)
        up3 = torch.cat((self.upsample3(up2), e3.clone()), dim=1)
        up4 = torch.cat((self.upsample4(up3), e2.clone()), dim=1)
        up5 = torch.cat((self.upsample5(up4), e1.clone()), dim=1)
        up6 = torch.cat((self.upsample6(up5), e0.clone()), dim=1)
        
        # last layers
        out = self.out(up6)
        out_act = self.act(out)
        return out_act
```

```python
generator = Generator().to(DEVICE)
summary(generator, input_size=(3, 256, 256))
```

```python
BCE = nn.BCEWithLogitsLoss()
L1 = nn.L1Loss()
```

```python
def generator_loss(dis_out, gen_out, target, Lambda=10000):
    gan_loss = BCE(dis_out, torch.ones_like(dis_out))
    l1_loss = L1(gen_out, target)
    total_loss = gan_loss + Lambda * l1_loss
    return total_loss
```

## Creating the Discriminator model

```python
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.enc_conv0 = nn.Sequential(
            nn.Conv2d(in_channels=3*2, out_channels=64, kernel_size=4, stride=2, padding=1, bias=False),
            nn.LeakyReLU()
        )
        self.enc_conv1 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU()
        )
        self.enc_conv2 = nn.Sequential(
            nn.Conv2d(in_channels=128, out_channels=256, kernel_size=4, stride=2, padding=3, bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU()
        )
        self.enc_conv3 = nn.Sequential(
            nn.Conv2d(in_channels=256, out_channels=512, kernel_size=4, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU()
        )
        self.out = nn.Conv2d(in_channels=512, out_channels=1, kernel_size=4, stride=1)
        
    def forward(self, x):
        #self.concated = torch.cat((input_img, target), dim=1)
        e0 = self.enc_conv0(x)
        e1 = self.enc_conv1(e0)
        e2 = self.enc_conv2(e1)
        e3 = self.enc_conv3(e2)
        out = self.out(e3)
        return out
```

```python
discriminator = Discriminator().to(DEVICE)
summary(discriminator, input_size=(6, 256, 256))
```

```python
def discriminator_loss(real, generated):
    real_loss = BCE(real, torch.ones_like(real))
    generated_loss = BCE(generated, torch.zeros_like(generated))
    total_loss = (real_loss + generated_loss) * 5
    return total_loss
```

## Weights initialization, optimizers creation

```python
# custom weights initialization called on netG and netD
def weights_init(model):
    for layer in model.children():
        if isinstance(layer, nn.Conv2d) or isinstance(layer, nn.ConvTranspose2d):
            nn.init.normal_(layer.weight.data, mean=0.0, std=0.02)
        if isinstance(layer, nn.BatchNorm2d):
            nn.init.normal_(layer.weight.data, mean=1.0, std=0.02)
            nn.init.constant_(layer.bias.data, 0)
```

```python
weights_init(generator)
weights_init(discriminator)
```

```python
optim_gen = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
optim_dis = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

```python
def show_images(model, input_img, target=None):
    model.eval()
    with torch.no_grad():
        pred = model(input_img)
    
    if target is None:
        plt.figure(figsize=(10, 10))
        imgs = [input_img, pred]
        title = ['Input', 'Predicted']
        for i in range(2):
            plt.subplot(1, 2, i + 1)
            plt.axis('off')
            plt.title(title[i])
            plt.imshow(prepare_image(imgs[i]))
        plt.show()
        return     
            
    plt.figure(figsize=(10, 20))
    imgs = [input_img, target, pred]
    title = ['Input', 'Target', 'Predicted']
    
    for i in range(3):
        plt.subplot(1, 3, i + 1)
        plt.axis('off')
        plt.title(title[i])
        plt.imshow(prepare_image(imgs[i]))
        
    plt.show()
```

## Learning

```python
def fit(gen, dis, epochs):
    
    gen_losses = []
    dis_losses = []
    
    print('Training has started...')
    
    for epoch in range(epochs):
        
        gen.train()
        dis.train()
        
        # training loop
        idx = random.sample(range(len(day_col)), len(day_col))
        
        # iterating over all images in dataset
        for index in idx:
            
            epoch_gen_losses = []
            epoch_dis_losses = []
            
            input_image, target_image = day_col[index], night_col[index]
            input_image, target_image = process_image(input_image, transformations), process_image(target_image, transformations)
            
            # unfreezing the discriminator
            for param in dis.parameters():
                param.requires_grad = True
            
            # calculating loss & updating weights for DISCRIMINATOR
            generated_image = gen(input_image)
            input_and_real_imgs = torch.cat((input_image, target_image), dim=1)
            input_and_generated_imgs = torch.cat((input_image, generated_image), dim=1)
            
            optim_dis.zero_grad()
            preds_for_generated = dis(input_and_generated_imgs.detach())
            preds_for_real = dis(input_and_real_imgs)
            
            dis_loss = discriminator_loss(preds_for_real, preds_for_generated)
            dis_loss.backward(retain_graph=True)
            optim_dis.step()
            
            # freezing the discriminator
            for param in dis.parameters():
                param.requires_grad = False
            
            # calculating loss & updating weights for GENERATOR
            optim_gen.zero_grad()
            preds_for_generated = dis(input_and_generated_imgs)
            gen_loss = generator_loss(preds_for_generated, generated_image, target_image)
            gen_loss.backward(retain_graph=True)
            optim_gen.step()
            
            epoch_gen_losses.append(gen_loss.item())
            epoch_dis_losses.append(dis_loss.item())
        
        gen_losses.append(np.mean(epoch_gen_losses))
        dis_losses.append(np.mean(epoch_dis_losses))
            
        # generating random image from test every epoch via generator
        display.clear_output(wait=True)
        ind = np.random.randint(0, len(day_col), 1).item()
        test_input = process_image(day_col[ind], transformations)
        test_real = process_image(night_col[ind], transformations)
        show_images(gen, test_input, test_real)
            
        print('[Epoch: %d/%d]\tLoss_G: %.4f\tLoss_D: %.4f'
            % (epoch + 1, epochs, gen_losses[-1], dis_losses[-1]))
```

```python
imsize = 256  

loader = transforms.Compose([
    transforms.Resize(imsize),
    transforms.CenterCrop(imsize),
    transforms.ToTensor()])
```

```python
class ContentLoss(nn.Module):

        def __init__(self, target,):
            super(ContentLoss, self).__init__()
            self.target = target.detach()
            self.loss = F.mse_loss(self.target, self.target)

        def forward(self, input):
            self.loss = F.mse_loss(input, self.target)
            return input
```

```python
def gram_matrix(input):
        batch_size, c, h, w = input.size()

        features = input.view(batch_size * c, -1)

        G = torch.mm(features, features.t())

        return G.div(batch_size * c * h * w)
```

```python
class StyleLoss(nn.Module):
        def __init__(self, target_feature, mask_img):
            super(StyleLoss, self).__init__()
            self.target = gram_matrix(target_feature).detach()
            self.mask = mask_img[0, 0, :, :]
            self.loss = F.mse_loss(self.target, self.target)

        def forward(self, input):
            n, c, h, w = input.size()
            curr_mask = np.resize(self.mask, (h, w))
            full_mask = np.zeros((n, c, h, w))
            for i in range(c):
                full_mask[0, i] = curr_mask
                
            full_mask = torch.from_numpy(full_mask).float().to(DEVICE)
            input = input * full_mask    
            
            G = gram_matrix(input)
            self.loss = F.mse_loss(G, self.target)
            return input
```

```python
cnn_normalization_mean = torch.tensor([0.485, 0.456, 0.406]).to(DEVICE)
cnn_normalization_std = torch.tensor([0.229, 0.224, 0.225]).to(DEVICE)
```

```python
class Normalization(nn.Module):
        def __init__(self, mean, std):
            super(Normalization, self).__init__()
            self.mean = torch.tensor(mean).view(-1, 1, 1)
            self.std = torch.tensor(std).view(-1, 1, 1)

        def forward(self, img):
            return (img - self.mean) / self.std
```

```python
def get_style_model_and_losses(cnn, normalization_mean, normalization_std,
                                   style_img, content_img, mask_img,
                                   content_layers=content_layers_default,
                                   style_layers=style_layers_default):
        cnn = copy.deepcopy(cnn)

        normalization = Normalization(normalization_mean, normalization_std).to(DEVICE)

        content_losses = []
        style_losses = []

        model = nn.Sequential(normalization)

        i = 0 
        for layer in cnn.children():
            if isinstance(layer, nn.Conv2d):
                i += 1
                name = 'conv_{}'.format(i)
            elif isinstance(layer, nn.ReLU):
                name = 'relu_{}'.format(i)
                # The in-place version doesn't play very nicely with the ContentLoss
                # and StyleLoss we insert below. So we replace with out-of-place
                # ones here.
                layer = nn.ReLU(inplace=False)
            elif isinstance(layer, nn.MaxPool2d):
                name = 'pool_{}'.format(i)
            elif isinstance(layer, nn.BatchNorm2d):
                name = 'bn_{}'.format(i)
            else:
                raise RuntimeError('Unrecognized layer: {}'.format(layer.__class__.__name__))

            model.add_module(name, layer)

            if name in content_layers:
                target = model(content_img).detach()
                content_loss = ContentLoss(target)
                model.add_module("content_loss_{}".format(i), content_loss)
                content_losses.append(content_loss)

            if name in style_layers:
                target_feature = model(style_img).detach()
                style_loss = StyleLoss(target_feature, mask_img)
                model.add_module("style_loss_{}".format(i), style_loss)
                style_losses.append(style_loss)

        # now we trim off the layers after the last content and style losses
        for i in range(len(model) - 1, -1, -1):
            if isinstance(model[i], ContentLoss) or isinstance(model[i], StyleLoss):
                break

        model = model[:(i + 1)]

        return model, style_losses, content_losses
```

```python
def get_input_optimizer(input_img):
        optimizer = optim.LBFGS([input_img.requires_grad_()]) 
        return optimizer
```

```python
def run_style_transfer(cnn, normalization_mean, normalization_std,
                        content_img, style_img, input_img, mask_img, num_steps=500,
                        style_weight=100000, content_weight=1):
    
        print('Building the style transfer model..')
        
        model, style_losses, content_losses = get_style_model_and_losses(cnn,
            normalization_mean, normalization_std, style_img, content_img, mask_img)
        optimizer = get_input_optimizer(input_img)

        print('Optimizing..')
        run = [0]
        while run[0] <= num_steps:

            def closure():
                input_img.data.clamp_(0, 1)

                optimizer.zero_grad()

                model(input_img)

                style_score = 0
                content_score = 0

                for sl in style_losses:
                    style_score += sl.loss
                for cl in content_losses:
                    content_score += cl.loss
                
                style_score *= style_weight
                content_score *= content_weight

                loss = style_score + content_score
                loss.backward()

                run[0] += 1
                if run[0] % 50 == 0:
                    print("run {}:".format(run))
                    print('Style Loss : {:4f} Content Loss: {:4f}'.format(
                        style_score.item(), content_score.item()))
                    print()

                return style_score + content_score

            optimizer.step(closure)

        input_img.data.clamp_(0, 1)

        return input_img
```
